## Coalign SMARP and SHARP maps

HMI and MDI observe the Sun from two different time-dependent locations, and both image in Heliprojective coordinates from their respective viewpoint. As a result, the same physical location on the disk of the Sun will not be located at the point in arcseconds. This notebook shows how address this problem and co-align magnetic field maps from HMI and MDI.

```python
import numpy as np
import matplotlib.pylab as plt
import sunpy.map
import drms
from astropy.io import fits
import matplotlib.colors as colors
from astropy.coordinates import SkyCoord
import astropy.units as u
from sunpy.visualization.colormaps import color_tables as ct
import sunpy.coordinates
from reproject import reproject_interp
```

### Set up the data query

The first step in querying for SMARP and SHARP data is to establish a connection to JSOC. This can be done with the `Client()` class. To do this, first [register your e-mail address with JSOC](http://jsoc.stanford.edu/ajax/register_email.html).

```python
email='mbobra@stanford.edu'
#email='your@email.address' # Put in a registered e-mail address
c = drms.Client(debug=True, verbose=True, email=email)
```

First we download the SHARP and SMARP maps, along with the full-disk HMI and MDI magnetic field maps.
We are querying data for a `T_REC` of 2010.10.15_19:12:00 for NOAA Active Region 11112, which corresponds to a `HARPNUM` of 211 and a `TARPUM` of 13643. We choose this example because the TARP and HARP look somewhat different for this active region.

```python
# Download the full-disk HMI magnetic field map
r = c.export('hmi.M_720s[2010.10.15_19:12:00]{magnetogram}', method='url', protocol='fits')
fits_url_hmifulldisk = r.urls['url'][0]
```

```python
# Download the full-disk MDI magnetic field map
q = c.export('mdi.fd_M_96m_lev182[2010.10.15_19:12:00]{data}', method='url', protocol='fits')
fits_url_mdifulldisk = q.urls['url'][0]
```

```python
# Download the SHARP magnetic field map
sharp = c.export('hmi.sharp_720s[211][2010.10.15_19:12:00]{magnetogram}', method='url', protocol='fits')
fits_url_sharp = sharp.urls['url'][0]
```

```python
smarp = c.export('mdi.smarp_96m[13643][2010.10.15_19:12:00]{magnetogram}', method='url', protocol='fits')
fits_url_smarp = smarp.urls['url'][0]
```

Reset FITS header keywords `CRDERn` from `nan` to 0.0.
Rotate maps if the FITS header keyword `CROTA2` is 180 (this means the image is upside-down).

```python
# For HMI full disk 
hmi_map = sunpy.map.Map(fits_url_hmifulldisk)
if hmi_map.meta['CRDER1'] == 'nan': hmi_map.meta['CRDER1'] = 0.0
if hmi_map.meta['CRDER2'] == 'nan': hmi_map.meta['CRDER2'] = 0.0

if int(hmi_map.meta['CROTA2']) == 180: 
    hmi_rotated = hmi_map.rotate(order=3)
else:
    hmi_rotated = hmi_map
```

```python
# For MDI full disk
mdi_map = sunpy.map.Map(fits_url_mdifulldisk)
if int(mdi_map.meta['CROTA2']) == 180: 
    mdi_rotated = mdi_map.rotate(order=3)
else:
    mdi_rotated = mdi_map
```

```python
# For SHARP
sharp_map = sunpy.map.Map(fits_url_sharp)
if sharp_map.meta['CRDER1'] == 'nan': sharp_map.meta['CRDER1'] = 0.0
if sharp_map.meta['CRDER2'] == 'nan': sharp_map.meta['CRDER2'] = 0.0
if int(sharp_map.meta['CROTA2']) == 180:
    sharp_rotated = sharp_map.rotate(order=3)
else:
    sharp_rotated = sharp_map
```

```python
# For SMARP
smarp_map = sunpy.map.Map(fits_url_smarp)
if int(smarp_map.meta['CROTA2']) == 180:
    smarp_rotated = smarp_map.rotate(order=3)
else:
    smarp_rotated = smarp_map
```

```python
fig = plt.figure(figsize=(15,10))

hmimag = plt.get_cmap('hmimag')

ax = plt.subplot(121, projection=sharp_rotated)
sharp_rotated.plot(cmap=hmimag, norm=colors.Normalize(vmin=-1500, vmax=1500))

ax = plt.subplot(122, projection=smarp_rotated)
smarp_rotated.plot(cmap=hmimag, norm=colors.Normalize(vmin=-1500, vmax=1500))
```

**Figure 1**: SHARP (left) and SMARP (right) magnetic field maps for NOAA Active Region 11112. These regions were observed on the on 2010-10-15 at 19:12:00 TAI. From these images it is clear that a given position (in arcseconds) in these images is not aligned.

---

While both HMI and MDI are in Helioprojective coordinates, these instruments have observed the Sun from different positions in space. There are two ways we can extract the same region.

First, we can co-align these images by extracting the submap using the the data from the other image.

### 1. Extract in projection

One way to coalign these data is to extract the HMI data using the SMARP coordinates, in the MDI reference frame.

```python
hmi_map_submap = hmi_rotated.submap(bottom_left=smarp_rotated.bottom_left_coord, 
                                    top_right=smarp_rotated.top_right_coord)
```

```python
fig = plt.figure(figsize=(15,10))
hmimag = plt.get_cmap('hmimag')

ax = plt.subplot(222, projection=sharp_rotated)
sharp_rotated.plot(cmap=hmimag, norm=colors.Normalize(vmin=-1500, vmax=1500))

ax = plt.subplot(221, projection=smarp_rotated)
smarp_rotated.plot(cmap=hmimag, norm=colors.Normalize(vmin=-1500, vmax=1500))

ax = plt.subplot(223, projection=smarp_rotated)
hmi_map_submap.plot(cmap=hmimag, norm=colors.Normalize(vmin=-1500, vmax=1500))
```

**Figure 1**: Top Left: SMARP region, which encapsulates NOAA Active Region 11112 on 2010-10-15 at 19:12:00 TAI. Top Right: HMI region extracted using the arcsecond values of the bounding box. This is an incorrect approach. The coordinates need to be converted between Helioprojective systems. Bottom Left: SMARP region extracted from HMI using `sunpy.submap`.

### 2. Reprojection

An alternative way is to reproject an image into another reference frame. Here we reproject HMI onto the MDI image. 
Using the AstroPy package called [reproject](https://reproject.readthedocs.io/en/stable/), we reproject HMI into the SMARP reference frame, and then extract the coordinates. In this case, both images will now be on the same grid ready to analyze together.

```python
output, footprint = reproject_interp(hmi_rotated, smarp_rotated.wcs, smarp_rotated.data.shape)
out_mdi = sunpy.map.Map(output, smarp_rotated.wcs)

oms = out_mdi.submap(bottom_left=out_mdi.bottom_left_coord, 
                         top_right=out_mdi.top_right_coord)
```

```python
fig = plt.figure(figsize=(15,10))
hmimag = plt.get_cmap('hmimag')

ax = plt.subplot(221, projection=sharp_rotated)
sharp_rotated.plot(cmap=hmimag, norm=colors.Normalize(vmin=-1500, vmax=1500))

ax = plt.subplot(223, projection=oms)
oms.plot(cmap=hmimag, norm=colors.Normalize(vmin=-1500, vmax=1500))

ax = plt.subplot(224, projection=smarp_rotated)
smarp_rotated.plot(cmap=hmimag, norm=colors.Normalize(vmin=-1500, vmax=1500))
```

**Figure 2**: Top Left: SHARP region, which encapsulates NOAA Active Region 11112 on 2010-10-15 at 19:12:00 TAI. Bottom row shows the HMI image reprojected on to the MDI reference frame. We use `.submap()` to extract the SMARP region (as right). Here we see that these images are now much better aligned.

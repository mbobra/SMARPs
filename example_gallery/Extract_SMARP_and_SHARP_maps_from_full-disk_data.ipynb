# Extract SMARP and SHARP maps from full-disk data

The SMARP and SHARP data provide partial-disk maps that encapsulate active regions. 

* For the data in Helioprojective Cartesian coordinates, we extract these partial-disk maps directly from the full-disk maps. 
* For the data in Heliographic Cylindrical Equal-Area coordinates, we extract these partial disk maps from the full-disk maps and then reproject the partial-disk maps in Cylindrical Equal-Area coordinates. Therefore, there is no corresponding full-disk image in Cylindrical Equal-Area coordinates.

This notebook specifically shows how to extract the SMARP and SHARP partial-disk maps in Helioprojective Cartesian coordinates directly from the full-disk maps.

```python
import sunpy.map
import drms
from astropy.io import fits
from sunpy.visualization.colormaps import color_tables as ct
import matplotlib.colors as colors
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as ptc
```

### Set up the data query

The first step in querying for SMARP and SHARP data is to establish a connection to JSOC. This can be done with the `Client()` class. To do this, first [register your e-mail address with JSOC](http://jsoc.stanford.edu/ajax/register_email.html).

```python
email='your@email.address' # Put in a registered e-mail address
c = drms.Client(debug=True, verbose=True, email=email)
```

The SMARP and SHARP data contain a bunch of segments, each corresponding to its own full-disk map.

**The SMARP data in Helioprojective Cartesian coordinates, available in `mdi.smarp_96m`**

| Segment | Description | Corresponding Full-Disk Series |
| ------- | ----------- | ------------------------------ |
| `bitmap` | A partial-disk merged bitmap that labels magnetic and photospheric activity (see the notebook Compare_SMARP_and_SHARP_bitmaps.ipynb for more information) | The full-disk magnetic activity bitmaps are in `mdi.fd_Marmask`, and the full-disk photospheric activity bitmaps are in `mdi.fd_spotmask`.|
| `magnetogram` | A partial disk line-of-sight magnetic field map | The full-disk magnetic field maps are in  `mdi.fd_M_96m_lev182`.|
| `continuum` | A partial-disk continuum intensity map | The full-disk continuum intensity maps are in `mdi.fd_Ic_interp`. |


**The SHARP data in Helioprojective Cartesian coordinates, available in `hmi.sharp_720s`**

Note: This series has 31 segments, most corresponding to outputs of the model that inverts the Stokes parameters to retrieve the magnetic field data. Here we will detail the segments both `hmi.sharp_720s` and `mdi.smarp_96m` have in common. For more information on the other segments, see [Bobra et al. (2014)](https://dx.doi.org/10.1007%2Fs11207-014-0529-3).

| Segment | Description | Corresponding Full-Disk Series |
| ------- | ----------- | ------------------------------ |
| bitmap | A partial-disk bitmap that labels magnetic activity (see the notebook Compare_SMARP_and_SHARP_bitmaps.ipynb for more information) | The full-disk magnetic activity bitmaps are in `hmi.Marmask_720s`.|
| magnetogram | A partial disk line-of-sight magnetic field map | The full-disk magnetic field maps are in  `hmi.M_720s`.|
| continuum | A partial-disk continuum intensity map | The full-disk continuum intensity maps are in `hmi.Ic_720s`. |

Query the full-disk magnetic field data:

```python
keys_mdi, segments_mdi = c.query('mdi.fd_M_96m_lev182[2010.07.14_11:12:00]', key=drms.const.all, seg='data')
url_mdi = 'http://jsoc.stanford.edu' + segments_mdi
mdi_fulldisk = fits.getdata(url_mdi.data[0])
```

Query the `magnetogram` segment of the SMARP data:

```python
keys_smarp, segments_smarp = c.query('mdi.smarp_96m[13520][2010.07.14_11:12:00]', key=drms.const.all, seg='magnetogram')
url_smarp = 'http://jsoc.stanford.edu' + segments_smarp.magnetogram[0]  
mdi_smarp = fits.getdata(url_smarp)
```

```python
XDIM_CCD = mdi_smarp.shape[1]
YDIM_CCD = mdi_smarp.shape[0]
print(f'The dimensions of this SMARP region are x: {XDIM_CCD} pixels, y: {YDIM_CCD} pixels.',
      f'The full-disk MDI image is x: {mdi_fulldisk.data.shape[1]} pixels, y: {mdi_fulldisk.data.shape[0]} pixels.')
```

### Visualize the SMARP data

```python
fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(1,1,1)
hmimag = plt.get_cmap('hmimag')
vmin=-2000
vmax=2000
plt.imshow(mdi_fulldisk, origin='lower', vmin=vmin, vmax=vmax, cmap=hmimag) # after rotation
```

```python
print(np.nanmax(mdi_fulldisk.data), np.nanmin(mdi_fulldisk.data))
```

```python
plt.imshow(mdi_smarp, origin='lower', vmin=vmin, vmax=vmax, cmap=hmimag)
```

```python
print(np.nanmax(mdi_smarp), np.nanmin(mdi_smarp))
```

We can identify this area on the full-disk photospheric magnetic field map for a sanity check:

```python
fig, ax = plt.subplots(1,1, figsize=(15,8))
plt.imshow(mdi_fulldisk,cmap=hmimag,origin='lower',vmin=-2000,vmax=2000)

y2 = (mdi_fulldisk.shape[0] - keys_mdi.CRPIX2 - keys_smarp.CRPIX2[0] + YDIM_CCD)  
y1 = (mdi_fulldisk.shape[0] - keys_mdi.CRPIX2 - keys_smarp.CRPIX2[0])
x2 = (mdi_fulldisk.shape[0] - keys_mdi.CRPIX1 - keys_smarp.CRPIX1[0] + XDIM_CCD)
x1 = (mdi_fulldisk.shape[0] - keys_mdi.CRPIX1 - keys_smarp.CRPIX1[0])

# draw a box using matplotlib.patches.Rectangle
ax.add_patch(ptc.Rectangle((x1, y1), XDIM_CCD, YDIM_CCD, hatch='\\\\\\\\', fill=False, snap=False))

# plot a white dot at the lower left-hand corner
plt.plot(x1, y1, 'w.', markersize=10.0)

# plot a black dot at the upper right-hand corner
plt.plot(x2, y2, 'k.', markersize=10.0)
```

Recalculate x1, x2, y1, and y2 as ints:

```python
y1 = int(np.rint(keys_smarp.IMCRPIX2[0] - keys_smarp.CRPIX2[0])) 
y2 = int(np.rint(keys_smarp.IMCRPIX2[0] - keys_smarp.CRPIX2[0] + YDIM_CCD))
x1 = int(np.rint(keys_smarp.IMCRPIX1[0] - keys_smarp.CRPIX1[0]))
x2 = int(np.rint(keys_smarp.IMCRPIX1[0] - keys_smarp.CRPIX1[0] + XDIM_CCD))
```

```python
print(x1,y1)
print(x2,y2)
```

Now extract the SMARP region from the full-disk data:

```python
subdata = mdi_fulldisk[y1:y2, x1:x2]
```

```python
fig = plt.figure(figsize=[20,20])
plt.imshow(subdata,cmap=hmimag,origin='lower',vmin=-2000,vmax=2000)
print('The dimensions of this image are',subdata.shape[0],'by',subdata.shape[1],'pixels.')
print('The max and min values of this image are',np.nanmax(subdata),'and',np.nanmin(subdata),'.')
```

Compare with the `magnetogram` segment of the SMARP data:

```python
fig = plt.figure(figsize=[20,20])
plt.imshow(mdi_smarp.data,cmap=hmimag,origin='lower',vmin=-2000,vmax=2000)
print('The dimensions of this image are',mdi_smarp.data.shape[0],'by',mdi_smarp.data.shape[1],'pixels.')
print('The max and min values of this image are',np.nanmax(mdi_smarp.data),'and',np.nanmin(mdi_smarp.data),'.')
```

We see that they are identically equal:

```python
difference_image = mdi_smarp.data-subdata
fig = plt.figure(figsize=[20,7])
plt.imshow(difference_image, origin='lower', cmap='RdBu', vmax=0.05, vmin=-0.05)
plt.colorbar()
np.mean(difference_image)
```

### Visualize the SHARP data

Go through the same process with SHARP data

```python
keys_hmi, segments_hmi = c.query('hmi.M_720s[2010.07.14_11:12:00]', key=drms.const.all, seg='magnetogram')
url_hmi = 'http://jsoc.stanford.edu' + segments_hmi
hmi_fulldisk = fits.getdata(url_hmi.magnetogram[0])
```

```python
keys_sharp, segments_sharp = c.query('hmi.sharp_720s[86][2010.07.14_11:12:00]', key=drms.const.all, seg='magnetogram')
url_sharp = 'http://jsoc.stanford.edu' + segments_sharp.magnetogram[0]  
hmi_sharp = fits.getdata(url_sharp)
```

```python
XDIM_CCD = hmi_sharp.data.shape[1]
YDIM_CCD = hmi_sharp.data.shape[0]
print(f'The dimensions of this SHARP region are x: {XDIM_CCD} pixels, y: {YDIM_CCD} pixels.',
      f'The full-disk HMI image is x: {hmi_fulldisk.shape[1]} pixels, y: {hmi_fulldisk.shape[0]} pixels.')
```

```python
y1 = int(np.rint(keys_sharp.IMCRPIX2[0] - keys_sharp.CRPIX2[0])) 
y2 = int(np.rint(keys_sharp.IMCRPIX2[0] - keys_sharp.CRPIX2[0] + YDIM_CCD))
x1 = int(np.rint(keys_sharp.IMCRPIX1[0] - keys_sharp.CRPIX1[0]))
x2 = int(np.rint(keys_sharp.IMCRPIX1[0] - keys_sharp.CRPIX1[0] + XDIM_CCD))
```

```python
subdata_hmi = hmi_fulldisk[y1:y2, x1:x2]
```

```python
fig = plt.figure(figsize=[20,20])
hmimag = plt.get_cmap('hmimag')
plt.imshow(subdata_hmi,cmap=hmimag,origin='lower',vmin=-2000,vmax=2000)
print('The dimensions of this image are',subdata_hmi.shape[0],'by',subdata_hmi.shape[1],'pixels.')
```

```python
fig = plt.figure(figsize=[20,20])
plt.imshow(hmi_sharp,cmap=hmimag,origin='lower',vmin=-2000,vmax=2000)
print('The dimensions of this image are',hmi_sharp.shape[0],'by',hmi_sharp.shape[1],'pixels.')
```

```python
difference_image_hmi = hmi_sharp-subdata_hmi
fig = plt.figure(figsize=[20,7])
plt.imshow(difference_image_hmi, origin='lower', cmap='RdBu')
plt.colorbar()
np.mean(difference_image_hmi)
```

```python
print(np.nanmax(difference_image_hmi), np.nanmin(difference_image_hmi))
```

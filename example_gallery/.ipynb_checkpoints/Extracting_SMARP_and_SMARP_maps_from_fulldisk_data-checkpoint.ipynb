### Data

There are five DRMS series that make up the SMARP workflow. Here they are for both HMI and MDI:

| Number | Purpose       | MDI  | HMI  |
| -- | ------------- | -----| -----|
| 1 | These are full-disk active region bitmaps based on line-of-sight magnetogram data. | `hmi.marmask_720s` | `mdi.fd_Marmask` |
| 2 | These are full-disk active region bitmaps based on continuum intensity data.       |   none |   `mdi.fd_spotmask` |
| 3 | These are partial-disk active region bitmaps. These bitmaps define membership into a coherent magnetic structure called an active region. For MDI, these bitmaps also include information about both the line-of-sight magnetic field and continuum intensity. For HMI, these bitmaps also include information about the line-of-sight magnetic field data.         |   `hmi.mharp_720s`  |  `mdi.mtarp` |
| 4 and 5| These are partial-disk active region cutouts. For MDI, these include cutouts of the full-disk line-of-sight magnetic field (available in `mdi.fd_M_96m_lev182` and full-disk temporally-interpolated continuum intensity (available in `mdi.fd_Ic_interp`), the partial-disk active region bitmap (available in `mdi.mtarp`), and active region summary parameters. For HMI, these include cutouts of the full-disk line-of-sight magnetic field (available in `hmi.M_720s`), full-disk continuum intensity data (available in `hmi.Ic_720s`), full-disk vector magnetic field data (available in `hmi.B_720s`), the partial-disk active region bitmap (available in `hmi.mharp_720s`), and active region summary parameters.      |   For CCD coordinates, see `hmi.sharp_720s`. And for CEA coordinates, see `hmi.sharp_cea_720s`. |  For CCD coordinates, see `su_mbobra.smarp_96m`. And for CEA coordinates, see `su_mbobra.smarp_cea_720s`. |


Query `mdi.fd_M_96m_lev182`:

```python
keys_mdi, segments_mdi = c.query('mdi.fd_M_96m_lev182[2010.10.15_00:00:00_TAI]', key=drms.const.all, seg='data')
url_mdi = 'http://jsoc.stanford.edu' + segments_mdi
mdi_fulldisk = fits.getdata(url_mdi.data[0])
```

Query a temporary SMARP series with `CDELTn` precision that matches that of the `mdi.fd_M_96m_lev182` series:

```python
keys_smarp, segments_smarp = c.query('su_mbobra.smarp_96m[13643][2010.10.15_00:00:00_TAI]', key=drms.const.all, seg='magnetogram')
url_smarp = 'http://jsoc.stanford.edu' + segments_smarp.magnetogram[0]  
mdi_smarp = fits.getdata(url_smarp)

XDIM_CCD = mdi_smarp.data.shape[1]
YDIM_CCD = mdi_smarp.data.shape[0]
print(f'The dimensions of this SMARP region are x: {XDIM_CCD} pixels, y: {YDIM_CCD} pixels.',
      f'The full-disk MDI image is x: {mdi_fulldisk.shape[1]} pixels, y: {mdi_fulldisk.shape[0]} pixels.')
```

Confirm that the `CDELTn` values in both series have the same precision:

```python
print(keys_smarp.CDELT1[0],keys_mdi.CDELT1[0])
print((keys_smarp.CDELT1[0])/(keys_mdi.CDELT1[0]))
print((keys_smarp.CDELT2[0])/(keys_mdi.CDELT2[0]))
ratio = ((keys_smarp.CDELT1[0])/(keys_mdi.CDELT1[0]))
print(f'The ratio of the SMARP:MDI platescales is {ratio}.')
```

We can identify this area on the full-disk photospheric image in CCD coordinates for a sanity check:

```python
fig, ax = plt.subplots(1,1, figsize=(15,8))
plt.imshow(mdi_fulldisk.data,cmap=hmimag,origin='lower',vmin=-2000,vmax=2000)

y2 = (mdi_fulldisk.shape[0] - keys_mdi.CRPIX2 - keys_smarp.CRPIX2[0] + YDIM_CCD)  
y1 = (mdi_fulldisk.shape[0] - keys_mdi.CRPIX2 - keys_smarp.CRPIX2[0])
x2 = (mdi_fulldisk.shape[0] - keys_mdi.CRPIX1 - keys_smarp.CRPIX1[0] + XDIM_CCD)
x1 = (mdi_fulldisk.shape[0] - keys_mdi.CRPIX1 - keys_smarp.CRPIX1[0])

# draw a box using matplotlib.patches.Rectangle
ax.add_patch(ptc.Rectangle((x1, y1), XDIM_CCD, YDIM_CCD, hatch='\\\\\\\\', fill=False, snap=False))

# plot a white dot at the lower left-hand corner
plt.plot(x1, y1, 'w.', markersize=10.0)

# plot a black dot at the upper right-hand corner
plt.plot(x2, y2, 'k.', markersize=10.0)
```

Recalculate x1, x2, y1, and y2 as ints:

```python
y1 = int(np.rint(keys_smarp.IMCRPIX2[0] - keys_smarp.CRPIX2[0]*(ratio))) 
y2 = int(np.rint(keys_smarp.IMCRPIX2[0] - keys_smarp.CRPIX2[0]*(ratio) + YDIM_CCD*(ratio)))
x1 = int(np.rint(keys_smarp.IMCRPIX1[0] - keys_smarp.CRPIX1[0]*(ratio)))
x2 = int(np.rint(keys_smarp.IMCRPIX1[0] - keys_smarp.CRPIX1[0]*(ratio) + XDIM_CCD*(ratio)))
```

```python
print(x1,y1)
print(x2,y2)
```

```python
subdata = mdi_fulldisk[y1:y2, x1:x2]
```

```python
fig = plt.figure(figsize=[20,20])
plt.imshow(subdata,cmap=hmimag,origin='lower',vmin=-2000,vmax=2000)
print('The dimensions of this image are',subdata.shape[0],'by',subdata.shape[1],'pixels.')
print('The max and min values of this image are',np.nanmax(subdata),'and',np.nanmin(subdata),'.')
```

```python
fig = plt.figure(figsize=[20,20])
plt.imshow(mdi_smarp,cmap=hmimag,origin='lower',vmin=-2000,vmax=2000)
print('The dimensions of this image are',mdi_smarp.shape[0],'by',mdi_smarp.shape[1],'pixels.')
print('The max and min values of this image are',np.nanmax(mdi_smarp),'and',np.nanmin(mdi_smarp),'.')
```

```python
difference_image = mdi_smarp-subdata
fig = plt.figure(figsize=[20,7])
plt.imshow(difference_image, origin='lower', cmap='RdBu')
plt.colorbar()
np.mean(difference_image)
```

```python
print(np.nanmax(difference_image), np.nanmin(difference_image))
```

### HMI sanity check

Go through the same process with HMI data for a sanity check

```python
keys_hmi, segments_hmi = c.query('hmi.M_720s[2010.10.15_00:00:00_TAI]', key=drms.const.all, seg='magnetogram')
url_hmi = 'http://jsoc.stanford.edu' + segments_hmi
hmi_fulldisk = fits.getdata(url_hmi.magnetogram[0])

keys_sharp, segments_sharp = c.query('hmi.sharp_720s[211][2010.10.15_00:00:00_TAI]', key=drms.const.all, seg='magnetogram')
url_sharp = 'http://jsoc.stanford.edu' + segments_sharp.magnetogram[0]  
hmi_sharp = fits.getdata(url_sharp)

ratio = ((keys_sharp.CDELT1[0])/(keys_hmi.CDELT1[0]))

XDIM_CCD = hmi_sharp.data.shape[1]
YDIM_CCD = hmi_sharp.data.shape[0]
print(f'The dimensions of this SHARP region are x: {XDIM_CCD} pixels, y: {YDIM_CCD} pixels.',
      f'The full-disk HMI image is x: {hmi_fulldisk.shape[1]} pixels, y: {hmi_fulldisk.shape[0]} pixels.')
y1 = int(np.rint(keys_sharp.IMCRPIX2[0] - keys_sharp.CRPIX2[0]*(ratio))) 
y2 = int(np.rint(keys_sharp.IMCRPIX2[0] - keys_sharp.CRPIX2[0]*(ratio) + YDIM_CCD*(ratio)))
x1 = int(np.rint(keys_sharp.IMCRPIX1[0] - keys_sharp.CRPIX1[0]*(ratio)))
x2 = int(np.rint(keys_sharp.IMCRPIX1[0] - keys_sharp.CRPIX1[0]*(ratio) + XDIM_CCD*(ratio)))
subdata_hmi = hmi_fulldisk[y1:y2, x1:x2]
```

```python
fig = plt.figure(figsize=[20,20])
hmimag = plt.get_cmap('hmimag')
plt.imshow(subdata_hmi,cmap=hmimag,origin='lower',vmin=-2000,vmax=2000)
print('The dimensions of this image are',subdata_hmi.shape[0],'by',subdata_hmi.shape[1],'pixels.')
```

```python
fig = plt.figure(figsize=[20,20])
plt.imshow(hmi_sharp,cmap=hmimag,origin='lower',vmin=-2000,vmax=2000)
print('The dimensions of this image are',hmi_sharp.shape[0],'by',hmi_sharp.shape[1],'pixels.')
```

```python
difference_image_hmi = hmi_sharp-subdata_hmi
fig = plt.figure(figsize=[20,7])
plt.imshow(difference_image_hmi, origin='lower', cmap='RdBu')
plt.colorbar()
np.mean(difference_image_hmi)
```

```python
print(np.nanmax(difference_image_hmi), np.nanmin(difference_image_hmi))
```

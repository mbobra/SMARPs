#### Query the data

The first step in querying for SMARP and SHARP data is to establish a connection to JSOC. This can be done with the `Client()` class.

```python
email = 'mbobra@stanford.edu'
c = drms.Client(debug=True, verbose=True, email=email)
```

To find all the SHARP data series, execute the following regular expression query:

```python
si = c.info('su_mbobra.smarp_cea_96m')
```

```python
t_start = '2010.10.15_19:12:00'
duration = '12m'
harpnum = '211'
tarpnum = '13643'
```

```python
# Query the HMI data for this single T_REC
hmi_query_string = 'hmi.sharp_cea_720s['+harpnum+']['+t_start+'/'+duration+'][? (QUALITY<65536) ?]'
keyshmi, segshmi = c.query(hmi_query_string, key=drms.const.all, seg='magnetogram, bitmap')
```

```python
# Query the MDI data for this single T_REC
mdi_query_string = 'su_mbobra.smarp_cea_96m['+tarpnum+']['+t_start+'/'+duration+'][? (QUALITY<262144) ?]'
keysmdi, segsmdi = c.query(mdi_query_string, key=drms.const.all, seg='magnetogram, bitmap, continuum')
```

```python
# Get the HMI data
mag_hmi = fits.open('http://jsoc.stanford.edu' + segshmi.magnetogram[0])
bit_hmi = fits.open('http://jsoc.stanford.edu' + segshmi.bitmap[0], uint8=True)
```

```python
# Get the MDI data
mag_mdi = fits.open('http://jsoc.stanford.edu' + segsmdi.magnetogram[0])
bit_mdi = fits.open('http://jsoc.stanford.edu' + segsmdi.bitmap[0], uint8=True)
con_mdi = fits.open('http://jsoc.stanford.edu' + segsmdi.continuum[0])
```

#### Plot the SHARP CEA data

```python
hmimag = plt.get_cmap('Greys')
plt.imshow(mag_hmi[1].data,cmap=hmimag,origin='lower', vmin=-800, vmax=800)
plt.title("HMI SHARP line-of-sight magnetic field data")
```

The SHARP bitmap has five unique values (Table A.2 of SHARP paper) because they describe two different characteristics: the location of the pixel (whether the pixel is off disk or a member of the patch) and the character of the magnetic field (quiet or active). Here are all the possible values:

|Value | Keyword        | Definition           |
|------| ----------------- |-----------------|
| 0 | `OFFDISK` | The pixel is located somewhere off the solar disk. |
| 1 | `QUIET` | The pixel is associated with weak line-of-sight magnetic field. |
| 2 | `ACTIVE` | The pixel is associated with strong line-of-sight magnetic field. |
| 32 | `ON_PATCH` | The pixel is inside the patch. |

Here are all the possible permutations:

|Value | Definition        |
|------| -----------------|
| 0 | The pixel is located somewhere off the solar disk (red). |
| 1 | The pixel is located outside the patch and associated with weak line-of-sight magnetic field (yellow). |
| 2 | The pixel is located outside the patch and associated with strong line-of-sight magnetic field (blue). |
| 33 | The pixel is located inside the patch and associated with weak line-of-sight magnetic field (orange). |
| 34 | The pixel is located inside the patch and associated with strong line-of-sight magnetic field (green). |

#### Plot the SMARP CEA data

```python
hmimag = plt.get_cmap('Greys')
plt.imshow(mag_mdi[1].data,cmap=hmimag,origin='lower', vmin=-800, vmax=800)
plt.title("MDI SMARP line-of-sight magnetic field data")
```

The SMARP bitmap has 13 unique values because they describe three different characteristics: the location of the pixel (whether the pixel is off disk or a member of the patch), the character of the magnetic field (quiet or active), and the character of the continuum intensity data (quiet, part of faculae, part of a sunspot). Here are all the possible values:

|Value | Keyword        | Definition           |
|------| ----------------- |-----------------|
| 0 | `OFFDISK` | The pixel is located somewhere off the solar disk. |
| 1 | `QUIET` | The pixel is associated with weak line-of-sight magnetic field. |
| 2 | `ACTIVE` | The pixel is associated with strong line-of-sight magnetic field. |
| 4 | `SPTQUIET` | The pixel is associated with no features in the continuum intensity data. |
| 8 | `SPTFACUL` | The pixel is associated with faculae in the continuum intensity data. |
| 16 | `SPTSPOT` | The pixel is associated with sunspots in the continuum intensity data. |
| 32 | `ON_PATCH` | The pixel is inside the patch. |

Here are all the possible permutations:

|Value | Definition        |

| 17 | The pixel is located outside the patch, associated with weak line-of-sight magnetic field, and sunspots in the continuum intensity data. |

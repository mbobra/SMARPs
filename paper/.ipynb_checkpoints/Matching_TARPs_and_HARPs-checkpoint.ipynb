# Matching HARPs and TARPs during the overlap period

The SMARP and SHARP data overlap for half a year, between 1 May 2010 and 28 October 2010. This overlap period provides an opportunity to characterize the differences between the two data sets. During this period, NOAA identified 57 active regions. The SMARP and SHARP data sets contain 127 and 137 active regions, respectively, during the same time period. In this notebook, we will match HARP regions with TARP regions during this overlap period.

```python
import numpy as np
import matplotlib.pylab as plt
from astropy.io import fits
import sunpy.map
from sunpy.visualization.colormaps import color_tables as ct
import matplotlib.colors as colors
import drms
import pandas as pd
from datetime import datetime as dt_obj
from dateutil.relativedelta import *
import astropy.units as u
import matplotlib.gridspec as gridspec
import seaborn as sns
#%matplotlib inline
pd.set_option('max_rows', 150) # print more than the default 10 rows of the pandas dataframe
%config InlineBackend.figure_format='retina'
```

```python
email='your@email.address' # Put in a registered e-mail address
c = drms.Client(debug=True, verbose=True, email=email)
```

```python
def parse_tai_string(tstr,datetime=True):
    year   = int(tstr[:4])
    month  = int(tstr[5:7])
    day    = int(tstr[8:10])
    hour   = int(tstr[11:13])
    minute = int(tstr[14:16])
    if datetime: return dt_obj(year,month,day,hour,minute)
    else: return year,month,day,hour,minute
```

Define the overlap period. The overlap period begins when HMI takes regular observations, on May 1, 2010. The overlap period ends on the last day of the mdi.mtarp series, on October 28, 2010. 

MDI continued to take some observations until April 11, 2011 -- and even turned back on from March 6-14, 2012 -- but gaps in the MDI data became more frequent after October 28, 2010. Therefore the SMARP series ends on October 28, 2010.

```python
t_start = '2010.05.01_00:00:00_TAI'
t_end = '2010.10.28_00:00:00_TAI'
```

```python
delta_days = parse_tai_string(t_end) - parse_tai_string(t_start)
```

```python
print("There are",delta_days.days,"in the overlap period.")
```

Query the MDI data. We find that there are 127 unique TARPs during the overlap period.

```python
# The 180d@2h designation queries the data once every two hours. 
# This reduces the time to execute the query, but does not miss any HARPs or TARPs (there are no HARPs or TARPs that live for shorter than 2 hours).
mdi_query_string = 'mdi.smarp_cea_96m[]['+t_start+'/'+str(delta_days.days)+'d@2h]'
print("This is the query:",mdi_query_string)
kmdi = c.query(mdi_query_string, key='TARPNUM')
```

```python
unique_tarps = kmdi['TARPNUM'].unique()
print("There are",len(kmdi['TARPNUM'].unique()),"unique TARPs during the overlap period.")
```

Query the HMI data. We find that there are 137 unique TARPs during the overlap period.

```python
hmi_query_string = 'hmi.sharp_cea_720s[]['+t_start+'/'+str(delta_days.days)+'d@2h]'
print("This is the query:",hmi_query_string)
khmi = c.query(hmi_query_string, key='HARPNUM')
```

```python
unique_harps = khmi['HARPNUM'].unique()
print("There are",len(unique_harps),"unique TARPs during the overlap period.")
```

### Match active regions across the SMARP and SHARP series

Sometimes a HARP or TARP is associated with a NOAA active region number, populated in the `NOAA_ARS` field. The easiest way to match active regions across the SHARP and SHARP series is to use the `NOAA_ARS` field. 

However, sometimes the `NOAA_ARS` field is not populated for some HARPNUMs or TARPNUMs that are, in fact, associated with an active region. For example, the value for `NOAA_ARS` is blank for TARPNUM 13389 because the NOAA SRS files show incorrect coordinates for the center of the active region. Since the HARP algorithm detected this as a small-to-medium sized region, those coordinates were sufficiently close enough to the center of the HARP for the code to associate the two. But since the TARP algorithm observed this as a small region, the coordinates were not close enough to the the center of the TARP to associate the two. The NOAA SRS files do contain mistakes. Furthermore, our threshold is low by design. Otherwise we risk mismatches. In cases like these we can match active regions across the two series by querying their coordinates.

First, grab all the HARPs during the overlap period. We will grab one record per HARP, which corresponds to the mid-point of its lifetime. The code block below will take ~3 minutes to run.

```python
%%capture
list_of_dataframes = []

for i in range(len(unique_harps)):

    print('Gathering data for HARPNUM', unique_harps[i], '.')
    
    # query all records for this harpnum
    hmi_query_string = 'hmi.sharp_cea_720s['+str(unique_harps[i])+'][][? (QUALITY<65536) ?]'
    khmi = c.query(hmi_query_string, key='T_REC')    
    
    # find midpoint time
    midpoint_harp_int = int(len(khmi)/2)
    midpoint_harp_time = khmi['T_REC'][midpoint_harp_int]
    
    # query SHARP data for the midpoint time
    hmi_query_string = 'hmi.sharp_cea_720s['+str(unique_harps[i])+']['+midpoint_harp_time+'][? (QUALITY<65536) ?]'
    df = c.query(hmi_query_string, key='T_REC,HARPNUM,NOAA_ARS,LAT_FWT,LON_FWT,LAT_MAX,LON_MAX,SIZE,AREA') 
    
    list_of_dataframes.append(df)
```

```python
df2 = pd.concat(list_of_dataframes).reset_index()
```

These are all the HARPs during the overlap period. There is one record per HARP, which corresponds to the mid-point of each HARP's lifetime.

```python
df2
```

Now we will go through each of these 137 HARPs and match them to TARPs using the Stonyhurst latitude and longitude coordinates in the keys `LAT_FWT` and `LON_FWT`. The following code block will take ~3 minutes to run.

```python
%%capture

df2['TARPNUM'] = 0

for i in range(len(df2)):

    print('Looking for TARPNUMs that match HARPNUM', unique_harps[i], '.')
    
    # Query all records for this tarpnum
    mdi_query_string = 'mdi.smarp_cea_96m[]['+df2['T_REC'][i]+'/3h][? (QUALITY<262144) ?]'
    kmdi = c.query(mdi_query_string, key='T_REC,TARPNUM,NOAA_ARS,LAT_FWT,LON_FWT,LAT_MAX,LON_MAX')
    
    # Match based on coordinates
    # If there is no SMARP data within 3 hours of the SHARP T_REC, 
    # skip this HARP and conclude that there is no matching TARP
    if len(kmdi)==0:
        continue
    
    # Determine time difference between SHARP and SMARP record, if any
    sharp_time = parse_tai_string(df2['T_REC'][i])
    smarp_time = parse_tai_string(kmdi['T_REC'][0])
    outtime = relativedelta(sharp_time, smarp_time).minutes
    
    # If a TARP is within 3 degrees of a HARP, count it as associated
    if len(kmdi)>1:
        for j in range(len(kmdi)):
            delta_latitude = df2['LAT_FWT'][i] - kmdi['LAT_FWT'][j]
            delta_longitude = df2['LON_FWT'][i] - kmdi['LON_FWT'][j]
            if (np.abs(delta_latitude) < 3.0 and np.abs(delta_longitude) < 3.0):
                #print("Found matching TARP:",kmdi['TARPNUM'][j])
                #print(delta_latitude, delta_longitude)
                df2['TARPNUM'][i] = kmdi['TARPNUM'][j]
                continue
```

Here is the list again. Now some HARPs are matched with TARPs.

```python
df2
```

Using coordinates is a reasonable way to match HARPs and TARPs, but we can see from the list above that we will cannot match every HARPs with a TARP this way. For example, NOAA Active Region 11117 is HARPNUM 226, but also TARPNUM 13658. However, the SMARP series did not have data within 3 hours of the midpoint T_REC identified for the SHARP series. 

```python
mdi_query_string = 'mdi.smarp_cea_96m[][2010.10.25_13:00:00_TAI/3h][? (QUALITY<262144) ?]'
kmdi = c.query(mdi_query_string, key='T_REC,TARPNUM,NOAA_ARS,LAT_FWT,LON_FWT,LAT_MAX,LON_MAX')
len(kmdi)
```

Another problem is that sometimes the `LAT_FWT` and `LON_FWT` keys are missing. In that case, we can try to match on the keyword `NOAA_ARS`. Here is an example for NOAA Active Region 11064. We see that it is TARPNUM 13364, and, from the table above, HARPNUM 2. But we could not match on coordinates because `LAT_FWT` and `LON_FWT` are NaNs in this case. 

```python
mdi_query_string = 'mdi.smarp_cea_96m[][2010.05.03_01:48:00_TAI/3h][? (QUALITY<262144) ?]'
kmdi = c.query(mdi_query_string, key='T_REC,TARPNUM,NOAA_ARS,LAT_FWT,LON_FWT,LAT_MAX,LON_MAX')
kmdi
```

Therefore, we can try to match HARPs to TARPs using the `NOAA_ARS` keyword.

```python
hmi_query_string = 'hmi.sharp_cea_720s[]['+t_start+'/180d@2h][? (NOAA_ARS <> \'MISSING\') ?]'
khmi = c.query(hmi_query_string, key='NOAA_ARS,HARPNUM')

mdi_query_string = 'mdi.smarp_cea_96m[]['+t_start+'/180d@2h][? (NOAA_ARS <> \'MISSING\') ?]'
kmdi = c.query(mdi_query_string, key='NOAA_ARS,TARPNUM')

khmi = khmi.drop(khmi.index[khmi['NOAA_ARS'] == 'MISSING']);
kmdi = kmdi.drop(kmdi.index[kmdi['NOAA_ARS'] == 'MISSING']);

keys = pd.merge(khmi, kmdi, on='NOAA_ARS')

keys = keys.drop_duplicates().reset_index(drop='True')
```

```python
for i in range(len(df2)):
    #print(i)
    if df2.iloc[i]['NOAA_ARS'] != 'MISSING':
        if df2.iloc[i]['TARPNUM'] == 0:
            if df2.iloc[i]['HARPNUM'] != 0:
                df2.iloc[i, df2.columns.get_loc('TARPNUM')] = keys[keys['NOAA_ARS'] == df2.iloc[1]['NOAA_ARS']]['TARPNUM'].values[0]
```

This method works to capture some more matches. For example, we now see that NOAA Active Region 11064 is correctly matched to TARPNUM 13394.

```python
# Dataframe of HARPs associated with TARPS
df3 = df2[df2['TARPNUM'] == 0].reset_index()
df3['TYPE'] = 'Unassociated'
```

```python
# Dataframe of HARPs unassociated with TARPs
df4 = df2[df2['TARPNUM'] != 0].reset_index()
df4['TYPE'] = 'Associated'
```

It is difficult to algorithmically associate HARPs and TARPs beyond this point.

We can try to match them up visually by eye by looking at movies of TARPs [here](http://jsoc.stanford.edu/data/mdi/mdi-tarp/mdi-browse/), and movies of HARPs [here](http://jsoc.stanford.edu/data/hmi/HARPs_movies/definitive/). We can also play with the 3-degree match criterion above, or take HARPs at points other than mid-point. This may improve the association. We see that the remaining unassociated HARPs are quite small in size, with an `AREA` of 860 micro-Hemispheres or less.

```python
print("The maximum area of an HARP unassociated with a TARP is",np.nanmax(df3['AREA']),".")
```

```python
df = {
    "Area (mH)": df3['AREA'].append(df4['AREA']), 
    "Type": df3['TYPE'].append(df4['TYPE'])
}
```

```python
sns.histplot(data=df, x=df['Area (mH)'], stat='count', kde=1, hue="Type")
plt.title("Matching HARPs with TARPs During the Overlap Period")
plt.xlabel("Area (mH)")
```

# Figure 2

Figure 2 plots the following keywords for three different data series:
* `USFLUXL`, the total unsigned flux (in Maxwells) derived from the line-of-sight magnetic field
* `R_VALUE`, the R parameter (as detailed in Schrijver+2007), which is a measure of the flux near the polarity inversion line, in Maxwells
* `MEANGBL`, the mean value of the line-of-sight field gradient in Gauss/Mm
* `CMASKL`, the number of pixels that contribute to the calculation of a keyword

In addition, panel b plots the following keywords:
* `USFLUX`, the total unsigned flux (in Maxwells) derived from the vertical field
* `ERRVF`, the error in `USFLUX`

The first step in querying for the SMARP and SHARP data is to establish a connection to JSOC. This can be done with the Client() class. (Register to download data from JSOC [here](http://jsoc.stanford.edu/ajax/register_email.html)).

```python
email='mbobra@stanford.edu' # Put in a registered e-mail address
c = drms.Client(debug=True, verbose=True, email=email)
```

```python
def parse_tai_string(tstr,datetime=True):
    year   = int(tstr[:4])
    month  = int(tstr[5:7])
    day    = int(tstr[8:10])
    hour   = int(tstr[11:13])
    minute = int(tstr[14:16])
    if datetime: return dt_obj(year,month,day,hour,minute)
    else: return year,month,day,hour,minute
```

Query the data for these T_RECs to plot the active region summary parameter values

```python
t_start = '2010.07.09_00:00:00_TAI'
t_end = '2010.07.20_00:00:00_TAI'
harpnum = '86'
tarpnum = '13520'
```

```python
hmi_query_string    = 'hmi.sharp_cea_720s['+harpnum+']['+t_start+'-'+t_end+'][? (QUALITY<65536) ?]'
keyshmi             = c.query(hmi_query_string, key=drms.const.all)

mdi_query_string    = 'mdi.smarp_cea_96m['+tarpnum+']['+t_start+'-'+t_end+'][? (QUALITY<262144) ?]'
keysmdi             = c.query(mdi_query_string, key=drms.const.all)
```

```python
t_rec_hmi = np.array([parse_tai_string(keyshmi.T_REC[i],datetime=True) for i in range(len(keyshmi.T_REC))])
```

```python
t_rec_mdi = np.array([parse_tai_string(keysmdi.T_REC[i],datetime=True) for i in range(len(keysmdi.T_REC))])
```

Calculate the errors in the active region summary parameter values. Running this function takes ~15 minutes for the SMARP data and ~3 hours for the SHARP data.

```python
def get_error_values_via_drms(t_start,t_end,region_number,hmi=True,mdi=True):

    """function: get_data

    This function reads the appropriate data and metadata using the SunPy affiliated package drms.
    """
    if hmi:
        goodqual = '65536'
        series = 'hmi.sharp_cea_720s'
    
    if mdi:
        goodqual = '262144'
        series = 'mdi.smarp_cea_96m'
        
    time_query_string  = series+'['+region_number+']['+t_start+'-'+t_end+'][? (QUALITY<'+goodqual+') ?]'
    time_keys   = c.query(time_query_string, key='T_REC')
    
    print(time_query_string)

    RPARAM_errors = []
    USFLUXL_errors = []
    MEANGBL_errors = []
    T_RECs = []
    
    print("This function will calculate errors for",str(len(time_keys['T_REC'])),"T_RECs. It will take some time.")
    
    for i in range(len(time_keys['T_REC'])):

        print(i)
        query_string  = series+'['+region_number+']['+time_keys['T_REC'][i]+']'
        print(query_string)
        keys, segs    = c.query(query_string, key=drms.const.all, seg='magnetogram,bitmap')
        
        # get fits header key information
        rsun_ref = keys['RSUN_REF']
        dsun_obs = keys['DSUN_OBS']
        rsun_obs = keys['RSUN_OBS']
        cdelt1   = keys['CDELT1']

        # Note that the value of CDELT1 in the CEA series is in units of degrees per pixel.
        # The following calculation converts CDELT1 into arcseconds.
        # Therefore the variable cdelt1_arcseconds is in units of arseconds per pixel.
        radsindeg = np.pi/180.
        cdelt1_arcsec = (math.atan((rsun_ref*cdelt1*radsindeg)/(dsun_obs)))*(1/radsindeg)*(3600.)

        # read data
        url = 'http://jsoc.stanford.edu' + segs['magnetogram'][0]
        magnetogram = fits.open(url)
        url = 'http://jsoc.stanford.edu' + segs['bitmap'][0]
        bitmap = fits.open(url)
        
        # get dimensions
        nx     = magnetogram[1].data.shape[1]
        ny     = magnetogram[1].data.shape[0]

        # Create an error array. Liu et al. (2020) [DOI: 10.1007/s11207-012-9976-x] determined
        # the median noise in the MDI one-minute full-disk magnetograms is 26.4 Mx cm^(âˆ’2) 
        # (see Figure 9). We will assume this noise value is homogeneous throughout the disk to
        # estimate the error in the keyword quantities.
        los_err = np.ndarray(shape=(ny,nx), dtype=float)
        los_err.fill(26.4)
        
        if mdi:
            Rparam, Rparam_err = calculate_smarpkeys.computeR(magnetogram[1].data, los_err, nx, ny, cdelt1_arcsec)
            mean_derivative_blos, mean_derivative_blos_err = calculate_smarpkeys.computeLOSderivative(magnetogram[1].data, los_err, nx, ny, bitmap[0].data, rsun_ref, rsun_obs, cdelt1_arcsec)
            mean_vf, mean_vf_err, count_mask  = calculate_smarpkeys.compute_abs_flux_los(magnetogram[1].data, los_err, bitmap[0].data, nx, ny, rsun_ref, rsun_obs, cdelt1_arcsec)

        if hmi:
            Rparam, Rparam_err = calculate_sharpkeys.computeR(magnetogram[1].data, los_err, nx, ny, cdelt1_arcsec)
            mean_derivative_blos, mean_derivative_blos_err = calculate_sharpkeys.computeLOSderivative(magnetogram[1].data, los_err, nx, ny, bitmap[0].data, rsun_ref, rsun_obs, cdelt1_arcsec)
            mean_vf, mean_vf_err, count_mask  = calculate_sharpkeys.compute_abs_flux_los(magnetogram[1].data, los_err, bitmap[0].data, nx, ny, rsun_ref, rsun_obs, cdelt1_arcsec)

        T_RECs.append(time_keys['T_REC'][i])
        RPARAM_errors.append(Rparam_err)
        MEANGBL_errors.append(mean_derivative_blos_err)
        USFLUXL_errors.append(mean_vf_err[0])
    
    return [T_RECs, RPARAM_errors, USFLUXL_errors, MEANGBL_errors] 
```

```python
T_RECs_smarp, RPARAM_errors_smarp, USFLUXL_errors_smarp, MEANGBL_errors_smarp = get_error_values_via_drms(t_start, t_end, tarpnum, mdi=True,hmi=False)
```

```python
T_RECs_sharp, RPARAM_errors_sharp, USFLUXL_errors_sharp, MEANGBL_errors_sharp = get_error_values_via_drms(t_start, t_end, harpnum,hmi=True,mdi=False)
```

Individual plots

```python
fig, ax = plt.subplots(figsize=(18,9))

### Define some style elements

# Font (use pdf rendering because eps rendering does not support Helvetica)
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = 'Helvetica'

# markers
palette = sns.color_palette()
marker_style4 = dict(linestyle='', markersize=16, fillstyle='full',color=palette[0], markeredgecolor=palette[0])
marker_style2 = dict(linestyle='', markersize=16, fillstyle='full',color=palette[1], markeredgecolor=palette[1])

#ticks
ax.tick_params(length=7.0, width=2, labelsize=24)

# spines
ax.spines['top'].set_linewidth(2)
ax.spines['right'].set_linewidth(2)
ax.spines['bottom'].set_linewidth(2)
ax.spines['left'].set_linewidth(2)

### Convert MEANGBL from Gauss/pixel to Gauss/Mm
# define some constants
radsindeg = np.pi/180.
munaught  = 0.0000012566370614

# calculate cdelt1 in arcseconds for both SHARP and SMARP data
cdelt1_arcsec_hmi = np.array([(math.atan((keyshmi['RSUN_REF'][i]*keyshmi['CDELT1'][i]*radsindeg)/(keyshmi['DSUN_OBS'][i])))*(1/radsindeg)*(3600.) for i in range(len(keyshmi.T_REC))])
cdelt1_arcsec_mdi = np.array([(math.atan((keysmdi['RSUN_REF'][i]*keysmdi['CDELT1'][i]*radsindeg)/(keysmdi['DSUN_OBS'][i])))*(1/radsindeg)*(3600.) for i in range(len(keysmdi.T_REC))])

# calculate factor for both SHARP and SMARP data
factor_hmi = np.array([((1/cdelt1_arcsec_hmi[i])*(keyshmi['RSUN_OBS'][0]/keyshmi['RSUN_REF'][0])*(1000000)) for i in range(len(keyshmi.T_REC))])
factor_mdi = np.array([((1/cdelt1_arcsec_mdi[i])*(keysmdi['RSUN_OBS'][0]/keysmdi['RSUN_REF'][0])*(1000000)) for i in range(len(keysmdi.T_REC))])

# multiply MEANGBL by factor
meangbl_hmi = np.array([(keyshmi.MEANGBL[i]*factor_hmi[i]) for i in range(len(keyshmi.T_REC))])
meangbl_mdi = np.array([(keysmdi.MEANGBL[i]*factor_mdi[i]) for i in range(len(keysmdi.T_REC))])

### Ascribe the data to the axes

ax.errorbar(t_rec_mdi, meangbl_mdi, yerr=MEANGBL_errors_smarp, fmt='-o', **marker_style4, label='SMARP keyword MEANGBL')
ax.errorbar(t_rec_hmi, meangbl_hmi, yerr=MEANGBL_errors_sharp, fmt='-o', **marker_style2, label='SHARP keyword MEANGBL')
#ax.plot(t_rec_mdi, meangbl_mdi,'o',**marker_style4, label='SMARP keyword MEANGBL')
#ax.plot(t_rec_hmi, meangbl_hmi,'o',**marker_style2, label='SHARP keyword MEANGBL')
plt.xlim([t_rec_hmi[0],t_rec_hmi[-1]])

### Label the axes, plot, and legend
ax.set_xlabel('Date', fontsize=28)
ax.set_ylabel('Gauss Mm$\mathregular{^{-1}}$', fontsize=28)
ax.xaxis.labelpad=15
ax.yaxis.labelpad=15
ax.set_title("Mean Line-of-Sight Field Gradient", fontsize=32, y=1.01)
plt.legend(frameon=False, loc='upper left', fontsize=28, handletextpad=-0.3)
ax.annotate('a', xy=(1, 1), xytext=(-35, -25), xycoords='axes fraction', color='black', fontsize='42',
             textcoords='offset pixels', horizontalalignment='right',verticalalignment='top')

### Save the figure
fig.tight_layout()
fig.savefig('Figure2a.pdf',transparent=False, bbox_inches='tight', dpi=300)
```
